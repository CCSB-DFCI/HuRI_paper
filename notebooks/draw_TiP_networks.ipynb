```python
from py2cytoscape.data.cyrest_client import CyRestClient
cy = CyRestClient()
cy.session.delete()

network_id_map = {}

for tissue,g in TiP_graphs.items():
    network = cy.network.create_from_igraph(g,name=tissue)
    cy.layout.apply(name='force-directed', network=network)

    TiP_style_obj = cy.style.create(name='TiP_style',original_style=TiP_style)
    cy.style.apply(TiP_style_obj,network=network)
    pdf = network.get_pdf()
    target = open(config.analysis_path + network_name + '/' + project + '/' + network_name + '_' + tissue + '.network.detail.pdf','wb')
    target.write(pdf)
    target.close()

    hairball_style_obj = cy.style.create(name='hairball_style',original_style=hairball_style)
    cy.style.apply(hairball_style_obj,network=network)
    pdf = network.get_pdf()
    target = open(config.analysis_path + network_name + '/' + project + '/' + network_name + '_' + tissue + '.network.hairball.pdf','wb')
    target.write(pdf)
    target.close()

    network_id_map[tissue] = network.get_id()
```
